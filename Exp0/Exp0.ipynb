## 2. Dataset Description

The Student Performance dataset contains academic and behavioral metrics for students, designed to predict and analyze academic performance. This dataset is used for practicing data preprocessing, statistical analysis, and visualization techniques.

### Dataset Overview:
- **Target Purpose:** Predictive analysis of student performance
- **Data Types:** Mix of numerical features
- **Primary Focus:** Hours studied, attendance, assignments, exam scores

### Features Description:
1. **Hours_Studied** (numerical): Number of hours dedicated to studying
2. **Attendance** (numerical): Attendance percentage in class
3. **Assignment_Score** (numerical): Score obtained on assignments
4. **Midterm_Score** (numerical): Score obtained in midterm examination
5. **Final_Score** (numerical): Score obtained in final examination (Target variable for analysis)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
sns.set_style('whitegrid')
```

```python
# Load the student performance dataset
try:
    df = pd.read_csv('student_performance.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Dataset file 'student_performance.csv' not found.")
    print("Please ensure the file is in the same directory as this notebook.")
    df = None
```

```python
if df is not None:
    print("Dataset Shape:", df.shape)
    print("\n" + "="*80)
    print("First Few Rows:")
    print(df.head())
    print("\n" + "="*80)
    print("Data Types:")
    print(df.dtypes)
    print("\n" + "="*80)
    print("Missing Values:")
    print(df.isnull().sum())
    print("\n" + "="*80)
    print("Statistical Summary:")
    print(df.describe())
```

```python
if df is not None:
    # 1. Load Final_Score as NumPy array
    final_scores = df['Final_Score'].to_numpy()
    print("Final Scores:", final_scores)
    print("\n" + "="*80)
```

```python
if df is not None:
    # 3. Perform Min-Max normalization
    min_val = np.min(final_scores)
    max_val = np.max(final_scores)
    normalized_scores = (final_scores - min_val) / (max_val - min_val)
    print("Normalized Scores:", normalized_scores)
```

```python
if df is not None:
    # 2. Compute mean, median, and standard deviation
    mean_score = np.mean(final_scores)
    median_score = np.median(final_scores)
    std_score = np.std(final_scores)
    print(f"Mean: {mean_score:.2f}, Median: {median_score:.2f}, Standard Deviation: {std_score:.2f}")
    print("="*80)
```

## 5. Exercise 2: Pandas Data Handling

```python
if df is not None:
    # 1. Check shape, columns, and missing values
    print("Shape of the dataset:", df.shape)
    print("Columns:", list(df.columns))
    print("Missing values per column:")
    print(df.isnull().sum())
    print("="*80)
```

```python
if df is not None:
    # 2. Create Performance label based on Final_Score
    # Categorize: Low (<60), Medium (60-80), High (>80)
    df_copy = df.copy()
    df_copy['Performance'] = pd.cut(df_copy['Final_Score'], bins=[0, 60, 80, 100], labels=['Low', 'Medium', 'High'])
    print("First 5 rows with Performance label:")
    print(df_copy.head())
    print("\n" + "="*80)
    print("Performance Distribution:")
    print(df_copy['Performance'].value_counts())
```

## 6. Exercise 3: Matplotlib Visualization

```python
if df is not None:
    # 1. Line plot: Hours_Studied vs Final_Score
    plt.figure(figsize=(8, 5))
    plt.plot(df['Hours_Studied'], df['Final_Score'], marker='o')
    plt.xlabel('Hours Studied')
    plt.ylabel('Final Score')
    plt.title('Hours Studied vs Final Score')
    plt.grid(True)
    plt.show()
```

```python
if df is not None:
    # 2. Histogram of Final_Score
    plt.figure(figsize=(8, 5))
    plt.hist(df['Final_Score'], bins=10, edgecolor='black')
    plt.xlabel('Final Score')
    plt.ylabel('Frequency')
    plt.title('Histogram of Final Score')
    plt.show()
```

## Exercise 4: Seaborn Visualization

```python
if df is not None:
    # 1. Scatter plot using seaborn
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x='Hours_Studied', y='Final_Score')
    plt.title('Scatter Plot: Hours Studied vs Final Score')
    plt.show()
```

```python
if df is not None:
    # 2. Heatmap for correlation analysis
    plt.figure(figsize=(8, 6))
    corr = df.select_dtypes(include=[np.number]).corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
# 2. Heatmap for correlation analysis
plt.figure(figsize=(8, 6))
corr = df.select_dtypes(include=[np.number]).corr()  # Only numeric columns
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
if df is not None:
    # 3. Boxplot for categorical analysis
    df_copy = df.copy()
    df_copy['Performance'] = pd.cut(df_copy['Final_Score'], bins=[0, 60, 80, 100], labels=['Low', 'Medium', 'High'])
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_copy, x='Performance', y='Final_Score')
    plt.title('Boxplot: Performance vs Final Score')
    plt.show()
```
